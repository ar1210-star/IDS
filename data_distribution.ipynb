```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"data\NF_ToN_IoT.csv")
```

```python
data.shape
```

```python
data["Attack"].value_counts()
```

```python
data["Attack"].value_counts().plot(
	kind="bar",
    figsize=(15, 20),
)
plt.title("Distribution of Attacks in the Dataset")
plt.xlabel("Attack Type")
plt.ylabel("Count")
plt.show()
```

```python
sample_data = data.sample(frac=0.02, random_state=42	)
```

```python
num_cols = sample_data.select_dtypes(include = ['int64', 'float64']).columns

num_cols = [c for c in num_cols
            if c!="Label"]
```

```python
dist_data = np.log1p(sample_data[num_cols])
```

```python
skew_values = dist_data.skew().sort_values(ascending=False)
```

```python
skew_values
```

```python
top_skewed = skew_values.head(14).index
top_skewed
```
